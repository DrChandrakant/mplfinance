```python
import mplfinance as mpf
import pandas as pd
```

Read in daily data for the S&P 500 from November of 2019

```python
df = pd.read_csv('data/SPY_20110701_20120630_Bollinger.csv',index_col=0,parse_dates=True)
```

# Bollinger Bands Indicator

Bollinger Bands Indicator Included in Mplfinance Core
<br>
Now plot() Takes Parameter called "indicators" basically which is list or list of dict
<br>
To Plot Bollinger Band indicator Use "kind" as "BBand"
<br>
To Define Period or Length Timeframe Required "length" = Period (Should Be int, Default is 21)
<br>
To Define Colors of Line define "colors" = list of color (In this 3 colors)
<br>
To Define Legend Labels a list of Names in "legend_label". If Not Define Legend Will Be Disabled
<br>

```python
bband=[
    {"kind": "BBand", "length": 21, "colors": ['red','green','blue'],'legend_label': ['Upper','Middle','Lower']}
]
```

```python
mpf.plot(
    df,
    type="candle",
    style='yahoo',
    volume=True,
    figscale =1.2,
    figsize = (46,26),
    indicators = bband,
)
```

# Exponential Moving Average and Simple Moving Averages

Exponential Moving Average Included in Mplfinance Core
<br>
Now plot() Takes Parameter called "indicators" basically which is list or list of dict
<br>
To Plot Exponential Moving Average Use "kind" as "ema"
<br>
To Define Period or Length Timeframe Required "length" = Period (Should Be int, Default is 7)
<br>
To Define Multiple EMA length should be tuple of Required EMA's
<br>
To Define Colors of Line define "colors" = list of color (Length Should Be Equal of EMA's)
<br>
To Define Legend Labels a list of Names in "legend_label". If Not Define Legend Will Be Disabled
<br>

```python
ema_multiple=[
    {"kind": "ema", "length": (5,7), 'colors': ['green','red'],'legend_label': ['Upper','Middle','Low'] },
]
```

```python
# Plot the candlestick chart
mpf.plot(
    df.tail(30),
    type="candle",
    style='yahoo',
    figscale =1.2,
    figsize = (46,26),
    indicators = ema_multiple,
)
```

```python
ema_single=[
    {"kind": "ema", "length": 6, 'colors': ['green'],'legend_label': ['EMA 6'] },
]
```

```python
# Plot the candlestick chart
mpf.plot(
    df.tail(30),
    type="candle",
    style='yahoo',
    figscale =1.2,
    figsize = (46,26),
    indicators = ema_single,
)
```

```python
sma_single=[
    {"kind": "sma", "length": 7, 'colors': ['green'],'legend_label': ['EMA 6'] },
]
```

```python
# Plot the candlestick chart
mpf.plot(
    df.tail(30),
    type="candle",
    style='yahoo',
    figscale =1.2,
    figsize = (46,26),
    indicators = sma_single,
)
```

```python
sma_multiple=[
    {"kind": "sma", "length": (5,7), 'colors': ['green','red'],'legend_label': ['EMA 5','EMA7'] },
]
```

```python
# Plot the candlestick chart
mpf.plot(
    df.tail(30),
    type="candle",
    style='yahoo',
    figscale =1.2,
    figsize = (46,26),
    indicators = sma_multiple,
)
```
