```python
import mplfinance as mpf
import pandas as pd
```

Read in daily data for the S&P 500 from November of 2019

```python
infile = 'data/t_data.csv'
df = pd.read_csv(infile, index_col=0, parse_dates=True)
```

```python
indicator=[
    {"kind": "BBand", "length": 21, "colors": ['red','green','blue'],'legend_label': ['Upper','Middle','Lower']}
]
```

```python
mpf.plot(
    df,
    type="candle",
    style='yahoo',
    volume=True,
    figscale =1.2,
    figsize = (46,26),
    indicators = indicator,
)
```
