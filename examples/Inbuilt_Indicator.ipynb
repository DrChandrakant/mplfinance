```python
import mplfinance as mpf
import pandas as pd
```

Read in daily data for the S&P 500 from November of 2019

```python
df = pd.read_csv('data/SPY_20110701_20120630_Bollinger.csv',index_col=0,parse_dates=True)
```

```python
bband=[
    {"kind": "BBand", "length": 21, "colors": ['red','green','blue'],'legend_label': ['Upper','Middle','Lower']}
]
```

```python
mpf.plot(
    df,
    type="candle",
    style='yahoo',
    volume=True,
    figscale =1.2,
    figsize = (46,26),
    indicators = bband,
)
```

```python
ema_multiple=[
    {"kind": "ema", "length": (5,7), 'colors': ['green','red'],'legend_label': ['Upper','Middle','Low'] },
]
```

```python
# Plot the candlestick chart
mpf.plot(
    df.tail(30),
    type="candle",
    style='yahoo',
    figscale =1.2,
    figsize = (46,26),
    indicators = ema_multiple,
)
```

```python
ema_single=[
    {"kind": "ema", "length": 6, 'colors': ['green'],'legend_label': ['EMA 6'] },
]
```

```python
# Plot the candlestick chart
mpf.plot(
    df.tail(30),
    type="candle",
    style='yahoo',
    figscale =1.2,
    figsize = (46,26),
    indicators = ema_single,
)
```
